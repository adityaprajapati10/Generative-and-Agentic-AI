```python
import tensorflow as tf
import numpy as np
```

```python
input = [1,2,3,4,5]
output = [2,4,6,8,10]
```

```python
x = np.array(input, dtype=float)
y = np.array(output, dtype=float)   
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(2, activation='relu', input_shape=[1]),
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01), loss='mean_squared_error')
```

```python
model.fit(x, y, epochs=100, verbose=1)
model.predict(np.array([9], dtype=float))
```
